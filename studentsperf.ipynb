```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('/content/StudentsPerformance (1).csv')  # Adjust file path if needed

# Define which columns to plot
subjects = ['math score', 'reading score', 'writing score']

# Bin ranges for histogram
bins = range(0, 101, 10)

# Loop through each subject and create a histogram
for subject in subjects:
    plt.figure(figsize=(8, 6))
    plt.hist(df[subject], bins=bins, edgecolor='black')
    plt.title(f'Distribution of {subject.title()}')
    plt.xlabel('Marks Range')
    plt.ylabel('Number of Students')
    plt.grid(True)
    plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("/content/StudentsPerformance (1).csv")


# Compute correlation only for numeric columns
corr_matrix = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```
